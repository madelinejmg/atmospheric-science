```python
from atmosphere import Atmosphere
import matplotlib.pyplot as plt
```

### Midlatitude Vertical Temperature Profile 
The vertical distribution of temperature for typical conditions in the Earth's atmosphere for the four atmospheres into layers (troposphere, stratosphere, mesosphere, and thermosphere)

```python
z, T, p = Atmosphere.integrate_pressure(z_top_m=70000, dz=100.0)

# Compare with isothermal
T_mean = T.mean()
p_iso = Atmosphere.isothermal_pressure(z, T_mean)

print("Mean T(0–70 km) =", T_mean)
print("Surface pressure =", p[0]/100, "hPa")
```

```python
km = z / 1000.0
plt.figure()
plt.plot(T, km)
plt.xlabel("Temperature (K)")
plt.ylabel("Height (km)")
plt.title("Piecewise-Linear Temperature Profile (0–70 km)")
plt.grid(True, alpha=0.3)

plt.figure()
plt.semilogy(p/100.0, km, label="US76 piecewise T(z)")
plt.semilogy(p_iso/100.0, km, "--", label=f"Isothermal (T̄={T_mean:.1f} K)")
plt.xlabel("Pressure (hPa)")
plt.ylabel("Height (km)")
plt.title("Pressure Profile on Log Scale")
plt.legend()
plt.grid(True, which="both", alpha=0.3)
plt.show()
```
